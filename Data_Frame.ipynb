## Oczyszczanie danych - dropna()

```python
frame3 = pd.read_csv('C:/Users/juleq/OneDrive/Desktop/pandas_kurs/course-files/Mcdonalds.csv')
```

```python
frame3.head()
```

```python
frame3.loc[1, 'Calories'] = np.NaN
frame3.loc[2, 'TotalFat'] = np.NaN
```

```python
frame3.head()
```

```python
frame3.dropna().head() #wiersze z wartościacmi NaN zostały usunięte
```

```python
frame3.dropna(how = 'all').head() #how = all - powoduje ze usuniete zostana tylko pozycje, ktore wszystkie wartosci maja NaN
```

```python
frame3.dropna(subset = ['TotalFat'], how = 'any').head() #polecenie subset wskazuje na nazwy kolumn na ktore polecenie
#ma popatrzec
#dlatego usuniety został wiersz 2, bo on ma w TotalFat wartość NaN
```

```python
frame3.dropna(axis = 'rows').head()
#parametr axis odpowiada na to czy analizowac wiersze czy kolumny, dla azixs = 0 - analizujemy wiersze
#natomiast dla axis = 1 - analiza kolumn, mozna tez wstawic teskt, np axis = 'rows' - wiersze
```

```python
frame3.dropna(axis = 1).head() #w 2 kolumnach był NaN, wieć zsotały całe usunięte
```

```python
frame3.dropna(inplace = True)#domyślne axis = 0, dlatego usuniete zostały wiersze: 1, 2
```

```python
frame3.head() #inplace domyślnie jest False, czyli operacje wykonywane są na kopii oryginalnego Frame
```
