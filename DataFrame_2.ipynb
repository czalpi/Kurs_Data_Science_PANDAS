## Ranking 
Ranking to nie tylko sortowanie danych, to rÃ³wnieÅ¼ przypisanie okreÅ›lonych miejsc w tym rankingu wszystkich obiektÃ³w, ktÃ³re wystÄ™pujÄ… w danych

```python
frame4 = pd.read_csv("C:/Users/juleq/OneDrive/Desktop/pandas_kurs/course-files/Mcdonalds.csv",
                    usecols = ['Item', 'Category', 'Serving Size', 'Calories', 'TotalFat'])
```

```python
frame4.head()
```

```python
frame4['Calories'].rank().head()#rank() domyÅ›lnie zwraca seriÄ™
```

Wady rankingu:
1. widzimy to w serii a nie w nowej kolumnie doÅ‚Ä…czonej do data frame
2. nie ma moÅ¼liwoÅ›ci zweryfikowania, czemu dany obiekt jest na tym konkretnym miejscu i czy dobrze sa ustawione, a to dlatego ze dane nie sa posortowane ze wzgledu an kalorie
3. Ranking zbudowaliÅ›my tak ze zaczyna sie od dan najmniej kalorycznych

```python
frame4['Calories Rank'] = frame4['Calories'].rank()
```

```python
frame4.head()
```

```python
frame4.sort_values(by = 'Calories').head(20)#jest troche rzeczy ktore maja 8 kalorii, dlatego zostaÅ‚o im przypisane po 8.5
```

```python
frame4['Calories Rank'] = frame4['Calories'].rank(ascending = False)#chcemy ustawiÄ‡ malejÄ…co
```

```python
frame4.sort_values(by = 'Calories', ascending = False).head(10)
```

```python
#po pozycji 7 nie ma 8 ani 9 jest za to 8.5 i to podwojnie
#Ranking to nie tlyko uporzadkowanie danych, ale bardziej nadanie numerow poszczegolnym elementom ze wzgledu na cos
```

```python
#3 najbardziej kaloryczne dania w menu
#metoda I
frame4.sort_values(by = 'Calories', ascending = False, inplace = True)
frame4.head(3)
```

```python
#metoda II - dedykowana do tego funkcja
frame4.nlargest(n = 3, columns = 'Calories')
```

```python
#3 najmniej kaloryczne dania w menu
frame4.tail(3)#moÅ¼na tak bo juÅ¼ sÄ… posortowane malejÄ…co
```

```python
frame4.nsmallest(n = 3, columns = 'Calories')#sÄ… to zupeÅ‚nie inne produkty niÅ¼ wyÅ¼ej, ale nadal poprawnie
```

```python
color = ['green', 'green', 'blue', 'blue', 'red', 'red', 'white']
size = ['S', 'S', 'S', 'M', 'M', 'M', 'L']
gender =  ['M', 'W', 'M', 'W', 'M', 'W', 'U']
clothes = pd.DataFrame({'color' : color, 'size' : size, 'gender' : gender})
clothes
```

```python
isColorBlue = clothes['color'] == 'blue'
isColorGreen = clothes['color'] == 'green'
isColorRed = clothes['color'] == 'red'
isGenderM = clothes['gender'] == 'M'
isGenderW = clothes['gender'] == 'W'
isSizeS = clothes['size'] == 'S'
isSizeM = clothes['size'] == 'M'
isSizeL = clothes['size'] == 'L'
```

```python
clothes[isGenderM & isColorBlue & isSizeS]#kolejnoÅ›Ä‡ podawania warunkow nie ma znaczenia
```

```python
clothes[isGenderM | isColorBlue | isSizeS]#nadal kolejnosc nie ma znaczenia, | - lub
```

```python
clothes[isGenderM & isColorBlue | isSizeM]#mnozenie jako pierwsze, wiec tutaj najpierw wykonywany jest warunek &
```

```python
clothes[isSizeM | isColorBlue & isGenderM]#zmiana kolejnoÅ›Ä‡i warunkÃ³w, ale bez zmiany operatorÃ³ miedzy nimi
```
